## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Set the bins and their names 
bins =[0,9,14,19,24,29,34,39,45]
group_lables=("<10","10-14","15-19","20-24","25-29","30-34","35-39","40+")
```

```python
#Categorize the playeres by age
age_g=pd.cut(purchase_data["Age"],bins,labels=group_lables)
age_g.head()
```

```python
# I use the same code the I have use for the gender but insted of filter by a column I  have use my age_g arange tu filter 
df_age=purchase_data.groupby(age_g)
age_d= df_age["SN"].nunique()
Total_age=pd.DataFrame(age_d)
Total_age.columns=["Total Count"]
```

```python
#same as the gender I add the percentages column
Total_age['Percentages'] = ((Total_age['Total Count']/576)*100).apply(lambda x: "{0:.2f}%".format((x)))
Total_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results

```python
df_sn=purchase_data.groupby(["SN"])
sn_d= df_sn["SN"].count()
sn_t=sn_d.sort_values()
top=sn_t.tail()
top_spe=pd.DataFrame(top)
top_spe.columns=["Purchase Count"]
top_spe
```
